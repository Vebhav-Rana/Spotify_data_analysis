FEATURE ENGINEERING

```python
df.index = range(0,len(df))
df
df["index"] = df.index
df
# fixing the index
```

```python
# making a new column as minutes
df["duration_min"] = df["duration_ms"]/60000
df
```

```python
# Making a different column for popularity_bucket
df["popularity_bucket"] = pd.cut(
    df["popularity"],
    bins=[-1,40,70,90,100],
    labels=["Low","Medium","High","Very High"]
)
```

```python
df["explicit"] = df["explicit"].replace({True:"Explicit",False:"Clean"})
df["mode"] = df["mode"].replace({1:"Major",0:"Minor"})

df
```

In this step, feature engineering was performed to improve interpretability of the dataset. The explicit column was converted from boolean values to human-readable labels, and the mode column was mapped from numeric encoding to major and minor scales. Track duration was converted from milliseconds to minutes for better readability. Additionally, a new feature popularity_bucket was created to generalize popularity scores into four categories: low, medium, high, and very high, enabling clearer comparative analysis.

```python
import matplotlib.pyplot as plt

top_artists = df["artists"].value_counts().head(10)

plt.figure()
top_artists.plot(kind="bar")
plt.title("Top 10 Artists by Number of Tracks")
plt.xlabel("Artist")
plt.ylabel("Track Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

**Insight:**  
A small number of artists contribute a large portion of tracks in the dataset, showing an uneven distribution of content.


```python
df.groupby("popularity_bucket")[[
    "energy",
    "danceability",
    "valence",
    "loudness",
    "duration_min"
]].mean()
```

```python
plt.figure()
plt.hist(df["popularity"], bins=20)
plt.title("Distribution of Track Popularity")
plt.xlabel("Popularity Score")
plt.ylabel("Number of Tracks")
plt.tight_layout()
plt.show()
```

**Insight:**  
Most tracks fall within the mid-range of popularity, with very few extremely high-popularity tracks.


The popularity distribution is highly skewed, with most songs falling into the low popularity category and only a small fraction achieving high or very high popularity. Analysis at the artist level suggests that having a larger number of songs does not necessarily correspond to higher average popularity, indicating that quantity alone is not strongly associated with popularity. Additionally, audio features such as energy, loudness, and danceability show noticeable variation across popularity levels. More popular songs also tend to have shorter durations on average compared to less popular tracks.

```python
top_20 = df.sort_values(by="popularity",ascending=False).head(int(0.2*len(df)))
bottom_20 = df.sort_values(by="popularity",ascending=False).tail(int(0.2*len(df)))
```

```python
top_20[["energy", "loudness", "danceability", "valence", "duration_min"]].mean()

```

```python
bottom_20[["energy", "loudness", "danceability", "valence", "duration_min"]].mean()
```

A comparison between the top 20% and bottom 20% of songs by popularity shows that highly popular songs tend to be louder, slightly more energetic, and marginally more danceable. Popular tracks are also shorter in duration on average. Valence does not show a positive relationship with popularity, indicating that emotional positivity alone is not a strong differentiator of popular songs.
